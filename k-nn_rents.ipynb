### Prepare Target

```python
def adjust_m2_price(df):
    df['m2_price'] = df['price'] / df['livingArea']
    m2_price_mean_germany = df['m2_price'].mean()

    m2_price_regional_supplement = {}
    for community in df['AP_community'].unique():
        m2_price_community_mean = df[df['AP_community'] == community]['m2_price'].mean()
        m2_price_regional_supplement[community] = m2_price_community_mean / m2_price_mean_germany

        df.loc[df['AP_community'] == community, 'm2_price'] = df.loc[df['AP_community'] == community, 'm2_price'] / m2_price_regional_supplement[community]

    del df['price']
    # del df['m2_price']
    return df, m2_price_regional_supplement


cursor_df, m2_price_regional_supplement = adjust_m2_price(cursor_df)

# prepare price category as alternative target
incrementor = cursor_df['m2_price'].min()
price_cat = 0
while incrementor < cursor_df['m2_price'].max():
    cursor_df.loc[cursor_df['m2_price'] >= incrementor, 'price_cat'] = price_cat
    print(price_cat, incrementor)
    incrementor *= 1.1
    price_cat += 1


cursor_df.describe(percentiles=[0.01, 0.1, 0.25, 0.5, 0.75, 0.90, 0.99])
```

### Preprocess

```python
def energyConsumption_to_int(x):
    if x == None:
        return x
    else:
        return float(x.split()[0])

cursor_df['energyConsumption'] = cursor_df['energyConsumption'].apply(lambda x: energyConsumption_to_int(x))
```

```python
import numpy as np
def condition_to_int(df):

    cats = ['', 'Altbau', 'ENTKERNT', 'ERSTBEZUG', 'NEUWERTIG',
       'Neubau', 'PROJEKTIERT', 'ROHBAU', 'TEIL_RENOVIERUNGSBED',
       'TEIL_VOLLRENOVIERT', 'TEIL_VOLLSANIERT']

    for cat in cats:
        df[cat] = df['condition'].apply(lambda x: 1 if x == cat else 0)

    # df['condition'] = pd.Categorical(df['condition'], categories=cats)
    # one_hot = pd.get_dummies(df['condition'], prefix='condition')
    # one_hot = pd.get_dummies(df['condition'])
    df = df.drop('condition', axis = 1)
    # df = df.join(one_hot)
    return df

cursor_df = condition_to_int(cursor_df)
cursor_df
```

```python
def fillnans_with_meadian(df):
    for col in df.columns:
        df[col] = df[col].fillna(df[col].median())
    return df

ap_community = cursor_df.pop('AP_community')
cursor_df = fillnans_with_meadian(cursor_df)
cursor_df.columns
```

### Categorization

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV

X_train, X_test, y_train, y_test = train_test_split(cursor_df, cat_price, test_size=0.2, random_state=42)

X_train = X_train.to_numpy()
X_test = X_test.to_numpy()
y_train = y_train.to_numpy()
y_test = y_test.to_numpy()

parameters = {
    'n_neighbors': [1, 6, 12],
    'weights':['distance', 'uniform'],
    'leaf_size': [2, 30, 60],
    'p': [1, 2],
}

neigh_class = KNeighborsClassifier()
knn_gridsearch = GridSearchCV(neigh_class, parameters, n_jobs=-1, verbose=4, return_train_score=True, scoring='accuracy')
knn_gridsearch.fit(X_train, y_train)

print('Best Params:', knn_gridsearch.best_params_)

y_pred = knn_gridsearch.predict(X_test)

validation_df = pd.DataFrame({'y_pred_class': y_pred, 'y_true_class': y_test})
validation_df['error_class'] = validation_df['y_pred_class'] - validation_df['y_true_class'] + 1
validation_df['valid_class'] = validation_df['error_class'].where(validation_df['error_class'] == 1, 0)

print(f"Pecent of accuratly rated: {validation_df['valid_class'].mean() * 100} %")
validation_df.describe(percentiles=[0.01, 0.05, 0.25, 0.5, 0.75, 0.95, 0.99])
#41.7596398512429 %

import pickle
import datetime as dt
import json

model_name = f"knn_class_acc_{str(round(validation_df['valid_class'].mean()*100, 2))}%_{dt.datetime.now().isoformat()}"
pickle.dump(knn_gridsearch, open(f"{model_name}.sav", 'wb'))
with open(f"{model_name}_params.json", 'w') as params_file:
    json.dump(knn_gridsearch.best_params_, params_file)
```

### Regression

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV

X_train, X_test, y_train, y_test = train_test_split(cursor_df, m2_price, test_size=0.2, random_state=42)

# delete outliers
percentile_05 = y_train.quantile(0.05)
percentile_95 = y_train.quantile(0.95)

X_train = X_train[y_train > percentile_05]
y_train = y_train[y_train > percentile_05]

X_train = X_train[y_train < percentile_95]
y_train = y_train[y_train < percentile_95]

# turn to numpy
X_train = X_train.to_numpy()
X_test = X_test.to_numpy()
y_train = y_train.to_numpy()
y_test = y_test.to_numpy()

parameters = {
    'n_neighbors': [12], #[1, 6, 12],
    'weights':['distance'], # 'uniform'],
    'leaf_size': [2], #[2, 30, 60],
    'p': [1], #[1, 2],
}

neigh_reg = KNeighborsRegressor()
knn_gridsearch = GridSearchCV(neigh_reg, parameters, n_jobs=-1, verbose=4, return_train_score=True, scoring='neg_mean_absolute_percentage_error')
knn_gridsearch.fit(X_train, y_train)

print('Best Params:', knn_gridsearch.best_params_)

y_pred = knn_gridsearch.predict(X_test)

validation_df = pd.DataFrame({'y_pred_reg': y_pred, 'y_true_reg': y_test})
validation_df['rel_error_reg'] = round(abs((validation_df['y_pred_reg'] - validation_df['y_true_reg']) / validation_df['y_true_reg']) * 100)
validation_df['valid_reg'] = validation_df['rel_error_reg'].apply(lambda x: 1 if x < 10 else 0)

print(f"Percent of accuratly rated listings: {validation_df['valid_reg'].mean() * 100} %")
validation_df.describe(percentiles=[0.01, 0.05, 0.25, 0.5, 0.75, 0.95, 0.99])
#60.04110393423371 % 
#Best Params: {'leaf_size': 2, 'n_neighbors': 12, 'p': 1, 'weights': 'distance'}

import pickle
import datetime as dt
import json

model_name = f"knn_reg_acc_{str(round(validation_df['valid_reg'].mean()*100, 2))}%_{dt.datetime.now().isoformat()}"
pickle.dump(knn_gridsearch, open(f"{model_name}.sav", 'wb'))
with open(f"{model_name}_params.json", 'w') as params_file:
    json.dump(knn_gridsearch.best_params_, params_file)
```
